```python
from typing import List
from array import array
```

```python
def initialize_array(path_to_input:str) -> List[array]:
    grid = []
    line = array('I')
    with open(path_to_input,"r") as f:
        for char in f.read():
            if not char == '\n':
                line.append(int(char))
            else:
                grid.append(line)
                line = array('I')
    return grid
```

```python
grid = initialize_array("input")
```

Pour chaque élément :
- Je vais vers le haut, si j'arrive à la fin c'est visible, sinon next :
- Je vais vers le bas, idem
- droite
- gauche

Il faut sélectionner intelligemment le premier côté à regarder :
- si je suis plus près du haut, commence par le haut, etc...

```python
size = (len(grid),len(grid[1]))
```

```python
size
```

```python
number_invisible = 0
for i in range(1,size[0]):
    for j in range(1,size[1]):
        tree_h = grid[i][j]
        
        left_invisible = False
        for k in range(j-1,-1,-1):
            tmp_h = grid[i][k]
            if tmp_h>=tree_h:
                left_invisible = True
                break
        
        top_invisible = False
        for k in range(i-1,-1,-1):
            tmp_h = grid[k][j]
            if tmp_h>=tree_h:
                top_invisible = True
                break
                
        right_invisible = False
        for k in range(j+1,size[1]):
            tmp_h = grid[i][k]
            if tmp_h>=tree_h:
                right_invisible = True
                break
                
        bottom_invisible = False
        for k in range(i+1,size[0]):
            tmp_h = grid[k][j]
            if tmp_h>=tree_h:
                bottom_invisible = True
                break
        
        if all([left_invisible,right_invisible,top_invisible,bottom_invisible]):
            number_invisible +=1
```
