```python
from typing import List
from array import array
```

```python
def initialize_array(path_to_input:str) -> List[array]:
    grid = []
    line = array('I')
    with open(path_to_input,"r") as f:
        for char in f.read():
            if not char == '\n':
                line.append(int(char))
            else:
                grid.append(line)
                line = array('I')
    return grid
```

```python
grid = initialize_array("input")
```

```python
size = (len(grid),len(grid[1]))


best_scenic = 0
for i in range(1,size[0]):
    for j in range(1,size[1]):
        tree_h = grid[i][j]
        
        temp_scen=1
        
        z=0
        for k in range(j-1,-1,-1):
            z+=1
            tmp_h = grid[i][k]
            if tmp_h>=tree_h:
                break
        temp_scen*=z
        
        
        z=0
        for k in range(i-1,-1,-1):
            z+=1
            tmp_h = grid[k][j]
            if tmp_h>=tree_h:
                break
        temp_scen*=z
                
        z=0
        for k in range(j+1,size[1]):
            z+=1
            tmp_h = grid[i][k]
            if tmp_h>=tree_h:
                break
        temp_scen*=z
            
        z=0
        for k in range(i+1,size[0]):
            z+=1
            tmp_h = grid[k][j]
            if tmp_h>=tree_h:
                break
        temp_scen*=z
        
        if temp_scen>best_scenic:
            best_scenic =temp_scen
        
```
